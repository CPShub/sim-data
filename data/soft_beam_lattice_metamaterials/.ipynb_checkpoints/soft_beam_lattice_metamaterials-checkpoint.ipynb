```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load data

The data is contained in the following folder of the repository.

```python
folder = 'data/'
```

Data is available for two cells, X and BCC, and for 8 simulation cases. 

```python
cells = ['X', 'BCC']
cases = ['uniaxial', 'biaxial', 'planar', 'volumetric', 'shear', 'test1', 'test2', 'test3']
```

The following function is used to get the required data for given cell and simulation case.

```python
def load_data(cell, case):
    data = np.loadtxt('%s%s_%s.txt' % (folder, cell, case))
    Fs = data[:, 0:9].reshape([-1, 3, 3])
    Ps = data[:, 9:18].reshape([-1, 3, 3])
    Ws = data[:, 18]
    return Fs, Ps, Ws
```

```python
Fs, Ps, Ws = load_data(cells[0], cases[0])
print(Fs.shape)
```

## 2. Display data

The following function loads data and displays it for given cell and case.

```python
def plot_data(cell, case):
    Fs, Ps, Ws = load_data(cell, case)
    
    if case != 'shear':
        xs = Fs[:, 0, 0]
    else:
        xs = Fs[:, 0, 1]
    
    fig, ax = plt.subplots(1, 3, figsize=(15,4))
    axx = ax[0]
    for i1 in range(3):
        for i2 in range(3):
            axx.plot(
                xs,
                Fs[:, i1, i2],
                label= '%i%i' % (i1+1, i2+1)
                )
    axx.legend()
    axx.set_ylabel('%s - %s - $F_{ij}$' % (cell, case))
    if case != 'shear':
        axx.set_xlabel('$F_{11}$')
    else:
        axx.set_xlabel('$F_{12}$')
    
    axx = ax[1]
    for i1 in range(3):
        for i2 in range(3):
            axx.plot(
                xs,
                Ps[:, i1, i2],
                label= '%i%i' % (i1+1, i2+1)
                )
    axx.legend()
    axx.set_ylabel('%s - %s - $P_{ij}$' % (cell, case))
    if case != 'shear':
        axx.set_xlabel('$F_{11}$')
    else:
        axx.set_xlabel('$F_{12}$')
        
    axx = ax[2]
    axx.plot(xs, Ws)
    axx.set_ylabel('%s - %s - $W$' % (cell, case))
    if case != 'shear':
        axx.set_xlabel('$F_{11}$')
    else:
        axx.set_xlabel('$F_{12}$')
    
    plt.tight_layout()
    plt.show()  
```

Display complete dataset for given cell.

```python
cell = cells[0] # X
for case in cases:
    print('***********************************************')
    print(case)
    plot_data(cell, case)
```
